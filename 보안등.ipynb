```python
import pandas as pd
data = pd.read_csv("C:/Users/dksxo/Downloads/경기도_성남시_보안등정보.csv",encoding="cp949")
import folium
```

```python
data
```

```python
data_sub = data[['위도', '경도']]
```

```python
data_sub = data_sub.dropna()
```

```python
data_sub
```

```python
base_latitude = data_sub['위도'].mean()
base_longitude = data_sub['경도'].mean()
```

```python
map = folium.Map(location=[base_latitude, base_longitude],
                 zoom_start=10)
```

```python
for n in data_sub.index:
    latitude = data['위도'][n]
    longitude = data['경도'][n]
    
    iframe = folium.IFrame('<b>' + data['보안등위치명'][n] + '</b><br><br>')
    popup = folium.Popup(iframe, min_width=300, max_width=300)
    folium.Marker(
        [latitude, longitude],
        popup = popup,
        tooltip = data['보안등위치명'][n]
    ).add_to(map)

# html로 저장
map.save('성남시보안등.html')
```
